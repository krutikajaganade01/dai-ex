```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("food_ordering_behavior_dataset.csv")
```

```python
print(df.head())
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
df.drop_duplicates(inplace= True)
```

```python
print(df.isnull().sum())
```

```python
 print(df.head())
```

```python


# High Value Order
df['high_value_order'] = (df['order_value'] > df['order_value'].median()).astype(int)
df
```

```python
# Rating Category
df['good_rating'] = (df['rating_given'] >= 4).astype(int)
df
```

```python
# Age Groups
df['age_group'] = pd.cut(df['age'],
                         bins=[18,25,35,45,60],
                         labels=['Adult','Young','Medain','Senior'])

print(df[['age','age_group']].head())
df
```

```python
city_encoded=pd.get_dummies(df["city"])
city_encoded
```

```python
# Combine City and Cuisine
df['city_cuisine'] = df['city'] + "_" + df['cuisine']

print(df[['city', 'cuisine', 'city_cuisine']].head())
```

```python
# Customer Score
df['customer_score'] = df['order_value'] * df['rating_given']

print(df[['order_value', 'rating_given', 'customer_score']].head())
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
numeric_cols=df.select_dtypes(include=['int64','float64']).columns
df[numeric_cols]=scaler.fit_transform(df[numeric_cols])

print(numeric_cols)
print(df.head())
```

```python
print("Dataset Shape:",df.shape)
```

```python
print(df.describe())
```

```python
city_counts = df['city'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(city_counts.index, city_counts.values)
plt.title("Top 10 Cities by Orders")
plt.xlabel("City")
plt.ylabel("Orders")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
repeat_orders = df['is_repeat_order'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(repeat_orders,
        labels=repeat_orders.index,
        autopct='%1.1f%%')
plt.title("Repeat Order Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['age'], df['order_value'])

plt.title("Age vs Order Value")
plt.xlabel("Age")
plt.ylabel("Order Value")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df['order_value'])

plt.title("Order Value Distribution")
plt.ylabel("Order Value")
plt.show()
```
